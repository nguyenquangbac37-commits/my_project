1 Yêu cầu. Cước phí Taxi Thông Minh. 
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, 
kiểu số nguyên, đơn vị VND). 
 
Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): - Xe 4 chỗ: 11,000 VND - Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: - Xe 4 chỗ: từ 0.8→20 km nh 13,000 VND/km; trên 20 km tính 10,500 VND/km - Xe 7 chỗ: từ 0.8→20 km nh 14,000 VND/km; trên 20 km tính 12,000 VND/km 
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND. 
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + 
chờ). 
5. Điều kiện đặc biệt: - Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. - Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).

```python
import math
def cuocphitaxi(car_type, start_hour, km, wait_minutes):
    #Bước 1 tính phí đi xe
    faremove = 0
    if car_type == 4:
        faremove = 11000
        if km > 0.8:
            if km < 20:
                faremove +=  (km -0.8) * 13000
            else:
                faremove += ( 20 - 0.8) * 13000 + (km -20) * 10500
    elif car_type == 7:
        faremove = 12000
        if km > 0.8:
            if km < 20:
                faremove += (km -0.8)* 14000
            else:
                faremove += (20 - 0.8) * 14000 + (km - 20) * 12000
    #tính phí chờ
    farewait = 0
    if wait_minutes > 5:
        blocks = math.ceil(wait_minutes / 5)
        farewait = (blocks - 1) * 2000
    #phụ phí ban đêm
    subtotal = faremove + farewait
    night_hours = [22, 23, 0, 1, 2, 3, 4, 5]
    is_night = start_hour in night_hours
    if is_night:
        subtotal *= 1.2
    if km > 30 and not is_night:
        subtotal *= 0.9
    total = int(subtotal)
    if total < 50000 and km > 5:
        total = 50000
    return total
car_type, start_hour, km, wait_minutes = map(int, input().split())
print(f'{cuocphitaxi(car_type, start_hour, km, wait_minutes)} VND')
```

16 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def tinhgoc(A, B, C):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    u1, v1 = x1 - x2, y1 - y2
    u2, v2 = x3 - x2, y3 - y2
    tichvohuong = u1 * u2 + v1 * v2
    ma_ba = math.sqrt(u1**2 + v1**2)
    ma_bc = math.sqrt(u2**2 + v2**2)
    if ma_ba == 0 or ma_bc == 0:
        return '3 điểm này không phân biệt'
    cos_theta = tichvohuong / ( ma_bc * ma_ba)
    cos_theta = max(-1.0, min(1.0, cos_theta))
    degree = math.degrees(math.acos(cos_theta))
    return round(degree, 2)
x1, y1 = map(float, input().split())
x2, y2 = map(float, input().split())
x3, y3 = map(float, input().split())
A = x1, y1
B = x2, y2
C = x3, y3
print(tinhgoc(A, B, C))
```

chấm động, ngược lại trả về 'neither'. - Đầu vào: chuỗi s. - Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. - Quy tắc: - Cho phép có dấu + hoặc - ở đầu. - Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). - Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). - Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.

```python
def detect_number_type(s):
    temp_s = s
    if s.startswith('+') or s.startwith('-'):
        temp_s = s[1:]
    if not temp_s:
        return 'neither'
    if '.' in temp_s:
        parts = temp_s.split('.')
        if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
            return 'float'
    return 'neither'    
```

Yêu cầu: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) trong đó q là thương nguyên và r là số dư sao 
cho dividend = divisorq + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %). - Đầu vào: số nguyên dividend, divisor (divisor khác 0). - Đầu ra: bộ hai số nguyên (q, r). - Ví dụ: divide_with_remainder(5, 3) -> (1, 2). 

```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    return (q, r)
a, b = map(int, input().split())
print(divide_with_remainder(a, b))
```

Yêu cầu: Viết hàm is_product_divisible_by_10(a, b) trả về True nếu a b chia hết cho 10, ngược lại trả về False. - Đầu vào: số nguyên a, b (có thể âm hoặc bằng 0). - Đầu ra: giá trị boolean. - Gợi ý: Một số chia hết cho 10 nếu chữ số tận cùng là 0 (tức là chia hết cho cả 2 và 5).

```python
def  is_product_divisible_by_10(a, b):
    if a % 10 == 0 and b % 10 == 0:
        return True
    return False
a, b = map(int, input().split())
print(is_product_divisible_by_10(a, b))
```

Viết hàm trả về tam giác vuông cân cao n bằng ký tự ''. 
Dòng thứ i (đánh số từ 1) có i dấu '', các dòng nối với nhau bằng ký tự xuống dòng. 
Ví dụ với n = 4 

```python
def tamgiac(n):
    for i in range(1, n+1):
        print(i * '*')
    return ''
print(tamgiac(4))
```

Viết hàm trả về chuỗi biểu diễn hình chữ nhật N x M bằng ký tự '', các hàng cách nhau bởi ký tự xuống dòng. 
Ví dụ, n=2, m=3 

```python
def hcn(n, m):
    for i in range(1, n+1):
        print('*' * m)
    return ''
n, m = map(int, input().split())
print(hcn(n, m))
```

Viết hàm chuyển một chữ cái in hoa thành chữ thường. Nếu đầu vào không phải đúng một ký tự, phát sinh ValueError.

```python
def chuyenkitu(n):
    a = ord(n)
    if 65<= a<= 90:
        return chr(a+32)
    elif 97<= a<= 120:
        return chr(a)
    else:
        return ValueError('kí tự không hợp lệ')
n = input()
print(chuyenkitu(n))
```

Viết hàm trả về ký tự đứng sau ký tự x n vị trí trong bảng chữ cái. 
Giữ nguyên chữ hoa/chữ thường và quay vòng trong phạm vi A..Z hoặc a..z. 

```python
def timkitu(x, n):
    a = ord(x)
    if 65<= a <= 90:
        vitrimoi = (a - 65 - n) % 26
        return chr(vitrimoi + 65)
    elif 97 <= a<=120:
        vitrimoi = (a - 97 - n) % 26
        return chr(vitrimoi + 97)
    else:
        return 'Ký tự không hợp lệ'
x = input()
n = int(input())
print(timkitu(x, n))
```

Viết hàm trả về True nếu a, b, c có thể là độ dài ba cạnh của một tam giác; ngược lại trả về False

Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể 
tạo thành tam giác.

```python
def tinhdientichtamgiac(a, b, c):
    if ( a + b < c or b + c < a or a + c < b) and ( a>0 and b > 0 and c > 0):
        p = (a + b + c) /2
        s = (p * ( p - a) * (p - b) * (p - c))**0.5
        return s
    else:
        return 'Ba cạnh không phải 1 tam giác hợp lệ'
```

Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1. 

```python
def giaithua(n):
    if n < 0:
        return 'Không hợp lệ'
    else: 
        if n == 0:
            return 1
        else:
            tich = 1
            for i in range(1, n + 1):
                tich *= i
            return tich
print(giaithua(3))
```

Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4).

```python
def kiemtraso(n):
    if n <= 1:
        return ''
    else:
        tong = 0
        for i in range(1, n):
            if i < n and i > round((n-1)/2, 2):
                tong +=1
        return tong
print(kiemtraso(9))
```
