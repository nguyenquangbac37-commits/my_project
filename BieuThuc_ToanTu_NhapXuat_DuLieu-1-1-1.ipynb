Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần
nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập
phân) theo định dạng đúng của output

```python
a, b = map(int, input('Nhập hai số nguyên vào đây').split())
a1 = a + b
a2 = a - b
a2 = a*b
a3 = a // b
a4 = a % b
a5 = a / b
print(a1, a2, a3, a4, f'{a5:.2f}')
```

đồng hồ có giá x USD (x nhập từ bàn phím). Bởi Amazon không vận chuyển trực tiếp về
Việt Nam, Đạt phải trả 10 USD cho phí vận chuyển từ một nước trung gian. Khi vận chuyển
về Việt Nam, Đạt được thông báo sẽ phải trả thêm 30% thuế xuất nhập cảnh và 10% thuế
VAT cho chiếc đồng hồ nói trên. Viết chương trình tính tổng số tiền mà Đạt phải trả để mua

Trong bài này, bạn được yêu cầu viết chương trình vẽ ra một tam giác cân bằng các dấu sao 
trên 3 dòng. Hãy kết hợp việc sử dụng khoảng trắng với các dấu sao để in ra tam giác này. 
Ví dụ, đoạn mã để in 2 dòng đầu của tam giác: 
print("" * "") 
print("" *** "") 
Yêu cầu: In ra màn hình tam giác có 3 dòng như sau: 
* 
*** 
*****

```python
n = int(input())
for i in range (0, n):
    print('*' * (2*i + 1))
```
