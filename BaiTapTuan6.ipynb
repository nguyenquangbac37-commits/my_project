Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng 
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.

```python
danhsach = list(map(int, input().split()))
for i in range(len(danhsach) - 1):
    for j in range(len(danhsach)-1, i, -1):
        if danhsach[i] == danhsach[j]:
            danhsach.pop(j)
print(danhsach)
```

W6A4 
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó 

```python
line = input().split()
result = {}
for item in line:
    if ':' in item:
        key, value = item.split(':', 1)
        if key in result:
            result[key].append(value)
        else:
            result[key] = [value]
print(result)
```

W6A5 
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
'zeros': ...} (đếm số >0, <0, =0) 

```python
danhsach = list(map(int, input().split()))
thongke = {'positives': 0, 'negatives': 0, 'zeros':0}
for i in danhsach:
    if i > 0:
        thongke['positives'] += 1
    elif i < 0:
        thongke['negatives'] += 1
    else:
        thongke['zeros'] += 1
print(thongke)
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
theo thứ tự từ điển tăng dần của key.  

```python
line1 = input().strip()
dict1 = {}
for i in line1:
    k, v = i.strip(':')
    dict1[k] = int(v)
line2 = input().strip()
dict2 = {}
for i in line2:
    k, v = i.strip(':')
    dict2[k] = int(v)
ketqua = dict1.copy()
for key, value in dict2.items():
    if key in ketqua:
        ketqua[key] += value
    else:
        ketqua[key] = value
keys = list(ketqua.keys())
n = len(keys)
for i in range(n):
    for j in range(n - i - 1):
        if keys[j] > keys[j+1]:
            keys[j], keys[j+1] = keys[j+1], keys[j]
dict3 = {}
for k in keys:
    dict3[k] = ketqua[k]
print(dict3) 
```

W6A10 
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
danhsach = list(map(int, input().split()))
k = int(input())
danhsach1 = []
for i in range(len(danhsach)):
    for j in range(i + 1, len(danhsach)):
        if danhsach[i] + danhsach[j] == k:
            if danhsach[i] > danhsach[j]:
                danhsach1.append((danhsach[j], danhsach[i]))
            else:
                danhsach1.append((danhsach[i], danhsach[j]))
for i in range(len(danhsach1)):
    for j in range(len(danhsach1)-1, i, -1):
        if danhsach1[i] == danhsach[j]:
            danhsach1.remove(danhsach[j])
print(danhsach1)
```

[Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong 
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ 
nhất trong số đó. 

```python
line = input().split()
danhsach = []
for item in line:
    danhsach.append(int(item))
if len(danhsach)==0:
    print('Nhập lại')
count = {}
for i in danhsach:
    if i in count:
        count[i] += 1
    else:
        count[i]=1
max = 0
tansuat = list(count.values())
for i in tansuat:
    if i > max:
        max = i
result = float('inf')
for x in count:
    if count[i] == max:
        if x < result:
            result = x
print(result)
```

[Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo 
một dictionary mới bằng cách đảo ngược key và value của dictionary ban 
đầu. In ra dictionary mới.

```python
line = input().split()
dict = {}
for i in line:
    k, v = i.split(':')
    dict[k] = v
dictdaonguoc = {}
for key, value in dict.items():
    dictdaonguoc[value] = key
print(dictdaonguoc)
```

[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary 
mới.

```python
line = input('Nhập danh sách vào đây').split()
k = int(input('Nhập giá trị k '))
dict = {}
for i in line:
    a, v = i.split(':')
    dict[a] = int(v)
dictmoi = {}
for key, value in dict.items():
    if dict[key] > k:
        dictmoi[key] = dict[key]
print(dictmoi)
```
