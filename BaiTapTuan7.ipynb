Viết hàm tìm kiếm nhị phân 
trên một danh sách đã 
được sắp xếp tăng dần. 
Trả về chỉ số của phần tử 
hoặc -1 nếu không tìm 
thấy. mẫu hàm: def 
binary_search(arr, left, 
right, target):

```python
def binary_search(list, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if list[mid] == target:
        return mid
    elif target < mid:
        return binary_search(list, left, mid - 1, target)
    else:
        return binary_search(list, mid + 1, right, target)
danhsach = []
while True:
    n = input('Nhập số theo thứ tự tăng dần vào danh sách nếu muốn kết thúc thì nhấn x')
    if n == 'x':
        break
    danhsach.append(int(n))
left = 0
right = len(danhsach) - 1
target = int(input('Nhập giá trị muốn tìm'))
ketqua = binary_search(danhsach, left, right, target)
print(ketqua)
```

```python
def count_occurrences(arr, x):
    solan = 0
    for i in range(len(arr)-1):
        if arr[i] == x:
            solan += 1
    return solan
arr = [1, 2, 3, 3, 4, 4, 3, 5, 3, 3, 3, 7]
x = 3
print(f'số lần xuất hiện của số {x} là: {count_occurrences(arr, x)}')
```

W7A3 
Viết hàm thực hiện thuật 
toán bubble sort, Selection 
Sort để sắp xếp danh sách 
theo thứ tự tăng dần. Sắp 
xếp nổi bọt (Bubble Sort);Sắp xếp chọn (Selection 
Sort)

```python
def Bubble_sort(list):
    for i in range(len(list)-1):
        for j in range(i+1, len(list)):
            if list[i] > list[j]:
                list[i], list[j] = list[j], list[i]
    return list
def selection_sort(list):
    n = len(list)
    for k in range(n-1):
        min_index = k
        for i in range(k + 1, len(list)):
            if list[k] < list[min_index]:
                min_index = i
        list[min_index], list[k] = list[k], list[min_index]
    return list
danhsach = [1, 7, 2, 6, 5, 9, 8, 3, 4]
print(Bubble_sort(danhsach))
print(selection_sort(danhsach))
```

Viết hàm tìm phần tử có số 
lần xuất hiện nhiều nhất 
trong danh sách. Nếu có 
nhiều phần tử cùng số lần, 
trả về phần tử xuất hiện 
trước. 

```python
def xuathien(lit):
    dic = {}
    for i in lit:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
    max_dic= 0
    for value in dic.values():
        if value > max_dic:
            max_dic = value
    for i in dic:
        if dic[i] == max_dic:
            return i
danhsach = [2, 3, 1, 2, 3, 4, 1]
print(xuathien(danhsach))
```

nguyên và một số X, đếm 
tất cả các cặp (a, b) sao

```python
def demcap(lst, X):
    socap = 0
    for i in range(len(lst) - 1):
        for j in range(i+1, len(lst)):
            if (lst[i] + lst[j]) % X == 0:
                socap += 1
    return socap
danhsach = [2, 3, 2, 3, 1, 4, 5, 6, -1, 3, 2, 3]
X = 5
print(demcap(danhsach, X))
```

W7A7 
Cho một danh sách các số 
nguyên nums, hãy tìm độ 
dài của dãy con tăng 
nghiêm ngặt dài nhất 
(không nhất thiết phải liền 
kề).

```python
def kiemtra(lst):
    if not lst:
        return 0
    n = len(lst)
    dp = [1] * n
    for i in range(1, n):
        for j in range(i):
            if lst[i] > lst[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    max_length = 0
    for x in dp:
        if x > max_length:
            max_length = x
    return max_length
danhsach = [10, 9, 2, 5, 3, 7, 101, 18]
print(kiemtra(danhsach))
```

W7A8 
Bạn được cung cấp: một 
danh sách các khoảng thời 
gian interval = [[l1, r1], [l2, 
r2], ...] và một danh sách 
các truy vấn queries = [q1, 
q2, ...]  
Với mỗi truy vấn q, hãy tìm 
độ dài nhỏ nhất của 
khoảng thời gian [li, ri] sao 
cho li ≤ q ≤ ri.  
Nếu không tồn tại khoảng 
thời gian nào như vậy, hãy 
trả về -1 cho truy vấn đó. 
solve in < O(m.n), hint: 
binary search in the 
queries list 

```python
def solve_pure_math(intervals, queries):
    intervals_with_len = []
    for i in range(len(intervals)):
        l, r = intervals[i]
        intervals_with_len.append([r - l + 1, l, r])
    intervals_with_len.sort()
    res = []
    for q in queries:
        found_val = -1
        for length, l, r in intervals_with_len:
            if l <= q <= r:
                found_val = length
                break 
        res.append(found_val)
    return res
intervals = [[1, 4], [2, 4], [3, 6], [4, 4]]
queries = [2, 3, 4, 5]
print(solve_pure_math(intervals, queries))
```

W7A9 
Cho một danh sách các số 
nums, với mỗi phần tử 
nums[i], hãy đếm xem có 
bao nhiêu phần tử bên phải 
nhỏ hơn nums[i].  
Trả về danh sách các số 
đếm. 
Efficient solution: using 
heap or binary search tree

```python
def demso(lst):
    dic = {}
    for i in range(len(lst)):
        for j in range(i+1, len(lst)):
            if lst[i] > lst[j]:
                if i not in dic:
                    dic[i] = 1
                else:
                    dic[i] += 1
        if i == len(lst)-1:
            dic[i] = 0
    danhsach = []
    for i in dic.values():
        danhsach.append(i)
    return danhsach
danhsach = [3, 2, 2, 6, 1] 
print(demso(danhsach))
```

Cho một danh sách các số 
nguyên nums và một số 
nguyên m, hãy tìm tổng 
mảng con lớn nhất theo 
modulo m.  
Tức là, với bất kỳ mảng 
con nums[i:j] nào, hãy tính 
sum(nums[i:j]) % m và trả 
về kết quả lớn nhất có thể. 
Gợi ý: 
Duy trì tổng tiền tố (prefix 
sum) mod m và sử dụng 
tìm kiếm nhị phân để tìm 
tổng tiền tố nhỏ nhất chỉ 
lớn hơn tổng hiện tại một 
chút (để tối đa hóa hiệu số 
bao quanh). 

```python
def tongmang(lst, m):
    danhsach = []
    for i in range(len(lst)-1):
        for j in range(i+1, len(lst)):
            tong = 0
            for a in range(i, j+1):
                tong += lst[a]
                ketqua = tong % m
                danhsach.append(ketqua)
    L = danhsach[0]
    for i in danhsach:
        if i > L:
            L = i
    return L
danhsach = [3, 3, 9, 9, 5]
m = 7
print(tongmang(danhsach,m))
##Chương trình này chưa tối ưu hóa các vòng for lông nhau phức tạp
```

```python
def tong_mang_thuan_toan(lst, m):
    n = len(lst)
    max_mod = 0
    
    # Bước 1: Tạo mảng Prefix Sum (Tổng tích lũy) - Thuần toán học
    # P[i] lưu tổng từ phần tử 0 đến i-1
    P = [0] * (n + 1)
    for i in range(n):
        P[i+1] = (P[i] + lst[i]) % m
        
    # Bước 2: Duyệt tất cả các cặp (i, j) để tìm hiệu số lớn nhất
    # Sum(i, j) mod m = (P[j] - P[i]) mod m
    for j in range(1, n + 1):
        for i in range(0, j):
            # Công thức hiệu số trong toán học Modulo
            current_mod = (P[j] - P[i] + m) % m
            
            if current_mod > max_mod:
                max_mod = current_mod
                
    return max_mod

danhsach = [3, 3, 9, 9, 5]
m = 7
print(f"Kết quả: {tong_mang_thuan_toan(danhsach, m)}")
```

Dãy hài hòa là một dãy 
thỏa mãn điều kiện giá trị 
lớn nhất và giá trị nhỏ nhất 
của dãy khác nhau đúng 
bằng \( 1 \). 
Viết hàm trả về độ dài của 
dãy hài hòa con dài nhất có 
thể có trong mảng giá trị 
nguyên \( arr \) có \( n \) 
phần tử 
Ví dụ: 
Input: 1 3 2 2 5 2 3 7 
Output: 5 - Dãy hài hòa con 
dài nhất là [3,2,2,2,3] 

```python
def dayhaihoa(lst):
    L = 0
    for i in range(len(lst)-1):
        tong = 1
        for j in range(i+1, len(lst)):
            if (lst[i] == lst[j] + 1) or (lst[i] == lst[j] - 1) or (lst[i] == lst[j]):
                tong += 1
            if L < tong:
                L = tong
    return L
danhsach =  [27, 20, 17, 32, 7, 49, 19, 52, 26, 1, 16, 4, 47, 
42, 49, 47, 35, 51, 4, 8, 40, 15, 45, 39, 12]
print(dayhaihoa(danhsach))
```
