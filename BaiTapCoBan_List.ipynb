### Bài 3: 
Tìm giá trị trung bình của các số nguyên dương trong một list.
Nếu không có số dương nào thì trả về "invalid"

```python
#Tuan
def Average_positive_numbers_of_list(lst):
    if len(lst) == 0:
        return 'invalid'
    else:
        a = len(lst)
        tong = 0
        for i in lst:
            tong += i
    tongtb = tong / a
    return round(tongtb, 2)
danhsach = list(map(int, input().split()))
print(Average_positive_numbers_of_list(danhsach))
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def move_zero_to_first(lst):
    if len(lst) == 0:
        return None
    else:
        lst1 = []
        lst2 = []
        for i in lst:
            if i == 0:
                lst1.append(i)
            else:
                lst2.append(i)
    lst3 = lst1 + lst2
    return lst3
danhsach = list(map(int, input().split()))
print(move_zero_to_first(danhsach))
```

### Bài 6: Xóa tất cả các số giống nhau trong list, giữ lại đúng 1 lần, Các số khác giữ nguyên vị trí trước 
[1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3] 


```python
def unique_array(lst):
    if len(lst) == 0:
        return None
    else:
        lst1 = []
        for i in lst:
            if i not in lst1:
                lst1.append(i)
    return lst1
danhsach = list(map(int, input().split()))
print(unique_array(danhsach))
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang

9 8 7 6 5 4 0: Giam

1 2 3 7 1 0 2: Vo danh

1 2 2 4 5 7: Vo danh

```python
danhsach = list(map(int, input().split()))
def daytang(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] > list[j]:
                list[i], list[j] = list[j], list[i]
    return list
def daygiam(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] < list[j]:
                list[i], list[j] = list[j], list[i]
    return list
def kiemtra(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] == list[j]:
                return True
    return False
if kiemtra(danhsach):
    print('Vô danh')
else:
    if danhsach == daytang(danhsach.copy()):
        print('Dãy tăng')
    elif danhsach == daygiam(danhsach.copy()):
        print('Dãy giảm')
    else:
        print('Vô danh')
```

### Bài 8: Kiểm tra một danh sách số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac
2 2 34 4 56 7 89 : không phải ziczac


```python
def is_ziczac(lst):
    for i in range(1, len(lst)-1):
        if  not (( (lst[i] > lst[i-1]) and (lst[i] > lst[i+1]) ) or ((lst[i]<lst[i-1]) and (lst[i]<lst[i+1]))):
            return False
    return True
danhsach = list(map(int, input().split()))
print(is_ziczac(danhsach))
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về None

```python
def find_pos_of_k(lst, k):
    lst1 = []
    for i in range(len(lst)):
        if lst[i] == k:
            lst1.append(i)
    if len(lst1) == 0:
        return -1
    else:
        return lst1
danhsach = list(map(int, input().split()))
k = int(input())
print(find_pos_of_k(danhsach, k))
```

### Bài 10: 
Nhập vào tên một người, in ra kí tự đầu tiên và kí tự cuối cùng của tên đó.
VD: "Nguyen Ai Quoc" => ["N", "c"]

```python
def theFirst_Last_character(s):
    for i in s:
        if i != ' ':
            a = i
            break
    for i in range(len(s)-1, -1, -1):
        if s[i] != ' ':
            b = s[i]
            break
    if a is None:
        return None
    else:
        return a, b
s = input()
print(theFirst_Last_character(s))

```

### Bài 11: 
Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào xuất hiện trước thì liệt kê trước
VD 1 2 1 2 1 1 1 1

output: 
1: 6
2: 2


```python
def list_number_in_lst(lst):
    dic = {}
    for i in range(len(lst)):
        tong = 1
        for j in range(i+1, len(lst)):
            if lst[i] == lst[j]:
                tong+=1
        if lst[i] not in dic:
            dic[lst[i]] = tong
    return dic
danhsach = list(map(int, input().split()))
print(list_number_in_lst(danhsach)) 
```

VD 2 1 2 1 2 1 1 1 1

output: dict 



```python
def list_number_in_lst_order(lst):
    dic = {}
    for i in range(len(lst)):
        tong = 1
        for j in range(i+1, len(lst)):
            if lst[i] == lst[j]:
                tong+=1
        if lst[i] not in dic:
            dic[lst[i]] = tong
    sorted_keys = sorted(dic.keys())
    dic1 = {}
    for key in sorted_keys:
        dic1[key] = dic[key]
    return dic1
 

    

    
assert list_number_in_lst_order([2, 3, 3, 1, 1, 2, 1, ]) == {2: 2, 3: 2, 1: 3}
assert list_number_in_lst_order([2, 3, 8, 8, 8, 3, 1, 1, 2, 1, 4, 3, 4, 5]) == {2: 2, 3: 3, 8: 3, 1: 3, 4: 2, 5: 1}

print("Dung roi")
```

### Bài 12:

output:

```python
def count_character_of_string(s):
    dic = {}
    for i in range(len(s)):
        tong = 1
        for j in range(i+1, len(s)):
            if s[i] == s[j]:
                tong+=1
        if s[i] not in dic:
            dic[s[i]] = tong
    sorted_keys = sorted(dic.keys())
    dic1 = {}
    for key in sorted_keys:
        dic1[key] = dic[key]
    return dic1
danhsach = input()
print(count_character_of_string(danhsach)) 
```
