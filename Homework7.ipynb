Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
def ymin(danhsach):
    dem = {}
    for x in danhsach:
        dem[x] = dem.get(x, 0) + 1
    y = 0
    for x, c in dem.items():
        if c < x:
            y += x
        else:
            y += (c-x)
    return y
arr = [2, 2, 2, 3, 3] 
# Số 2: c=3, x=2 => xóa 3-2 = 1
# Số 3: c=2, x=3 => xóa 2
# Tổng y = 1 + 2 = 3
print(ymin(arr))


```

Nếu không có phần tử nào thỏa mãn thì không in gì 

```python
def dem(danhsach):
    n = len(danhsach)
    if n == 0:
        return
    candidate = None
    count = 0
    for i in danhsach:
        if count == 0:
            candidate = i
            count = 1
        elif i == candidate:
            count += 1
        else:
            count -= 1
    demlai = 0
    for i in danhsach:
        if i == candidate:
            demlai += 1
    if demlai > n //2:
        return candidate
danhsach = [2, 2, 1, 1, 1, 2, 2]
print(dem(danhsach))
#Sử dụng tính đối kháng chia các cặp đấu và tự triệt tiêu nếu chúng khác nhau, nếu giống nhau thì chúng sẽ cộng dồi lại
    
```

Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không 
dùng bất cứ thuật toán sắp xếp nào 

```python
def sapxep(lst1, lst2):
    def xemxet(danhsach):
        if len(danhsach)<=1:
            return 0, len(danhsach), 1
        if danhsach[0] > danhsach[-1]:
            return len(danhsach)-1, -1, -1
        return 0, len(danhsach), 1
    p1, end1, step1 = xemxet(lst1)
    p2, end2, step2 = xemxet(lst2)
    danhsachmoi = []
    while p1 != end1 and p2 != end2:
        if lst1[p1] <= lst2[p2]:
            danhsachmoi.append(lst1[p1])
            p1 += step1
        else:
            danhsachmoi.append(lst2[p2])
            p2 += step2
    while p1 != end1:
        danhsachmoi.append(lst1(p1))
        p1 += step1
    while p2 != end2:
        danhsachmoi.append(lst2[p2])
        p2 += step2
    return danhsachmoi
array_a = [1, 3, 5, 7]      # Tăng dần
array_b = [8, 6, 4, 2, 0]   # Giảm dần

print(f"Mảng sau khi trộn: {sapxep(array_a, array_b)}")
# Kết quả: [0, 1, 2, 3, 4, 5, 6, 7, 8]
```

W7A22 
Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. 
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
def tim_sinh_vien_hay_gap_nhat():
    # Nhập số lượng sinh viên N
    n = int(input())
    
    counts = {}
    
    # Bước 1: Nhập tên và đếm tần suất
    for _ in range(n):
        name = input().strip()
        counts[name] = counts.get(name, 0) + 1
    
    # Bước 2 & 3: Tìm tên xuất hiện nhiều nhất và nhỏ nhất về từ điển
    max_freq = -1
    result_name = ""
    
    for name, freq in counts.items():
        # Nếu tìm thấy tần suất lớn hơn hẳn
        if freq > max_freq:
            max_freq = freq
            result_name = name
        # Nếu tần suất bằng nhau, chọn tên nhỏ hơn theo từ điển
        elif freq == max_freq:
            if name < result_name:
                result_name = name
                
    return result_name

# Chạy chương trình
print(tim_sinh_vien_hay_gap_nhat())
```

W7A23 
Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài 
>=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí 
tự cái. 

```python
import re

def dem_so_luong_ten(text):
    # Bước 1: Tách chuỗi thành các từ. 
    # Ở đây dùng Regex để chỉ lấy các cụm chữ cái liên tiếp, loại bỏ dấu câu/số.
    words = re.findall(r'[a-zA-Z]+', text)
    
    count = 0
    for word in words:
        # Bước 2: Kiểm tra từng từ theo điều kiện
        # 1. Độ dài >= 2
        if len(word) >= 2:
            # 2. Chữ đầu viết hoa VÀ phần còn lại viết thường
            # word[0] là chữ đầu, word[1:] là phần còn lại
            if word[0].isupper() and word[1:].islower():
                count += 1
                
    return count

# --- Chạy thử ---
test_string = "Nguyen Van A di choi voi Tran Thi B12 và Alice."
# Giải thích: 
# - Nguyen: Thỏa mãn
# - Van: Thỏa mãn
# - A: Loại (độ dài = 1)
# - B12: Loại (có số 12, Regex đã tách thành "B" - độ dài 1)
# - Alice: Thỏa mãn
print(f"Số lượng tên tìm được là: {dem_so_luong_ten(test_string)}")
```

W7A24 Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên

```python
def xoa_tu_lap_lien_ke(text):
    # Bước 1: Tách xâu thành danh sách các từ (mặc định tách theo khoảng trắng)
    words = text.split()
    
    # Nếu xâu rỗng thì trả về xâu rỗng luôn
    if not words:
        return ""
    
    # Bước 2: Tạo danh sách kết quả, mặc định giữ lời từ đầu tiên
    result = [words[0]]
    
    # Bước 3: Duyệt từ từ thứ hai (vị trí index 1) đến hết
    for i in range(1, len(words)):
        # So sánh từ hiện tại với từ ngay phía trước nó
        if words[i] != words[i-1]:
            result.append(words[i])
            
    # Bước 4: Nối các từ lại thành xâu hoàn chỉnh bằng dấu cách
    return " ".join(result)

# --- Chạy thử ---
input_str = "tôi tôi đi đi học học học ở ở trường"
# Giải thích: 
# "tôi" (đầu tiên) -> giữ
# "tôi" (thứ hai) == "tôi" -> xóa
# "đi" (thứ nhất) != "tôi" -> giữ
# ...
output_str = xoa_tu_lap_lien_ke(input_str)
print(f"Kết quả: '{output_str}'")
```

W7A25 
Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi 
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số 

```python
def tim_so_lon_nhat(s, k):
    n = len(s)
    so_luot_xoa = n - k
    stack = []
    
    for digit in s:
        # Trong khi còn lượt xóa và số hiện tại 'ngầu' hơn số trong stack
        while so_luot_xoa > 0 and stack and stack[-1] < digit:
            stack.pop()
            so_luot_xoa -= 1
        stack.append(digit)
    
    # Chỉ lấy đúng k chữ số đầu tiên (phòng trường hợp chưa xóa hết)
    result = "".join(stack[:k])
    return result

# --- Chạy thử ---
# n = 7 chữ số, k = 4. Được phép xóa 3 số.
s = "4325671" 
k = 4
print(f"Số lớn nhất có {k} chữ số là: {tim_so_lon_nhat(s, k)}")
```

W7A26 
Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các 
số đó.

```python
import re

def xu_ly_so_trong_van_ban(text):
    # Bước 1: Tìm tất cả các cụm chữ số trong văn bản
    # \d+ nghĩa là tìm một hoặc nhiều chữ số liên tiếp
    danh_sach_chuoi_so = re.findall(r'\d+', text)
    
    # Bước 2: Chuyển danh sách chuỗi thành danh sách số nguyên
    danh_sach_so = [int(x) for x in danh_sach_chuoi_so]
    
    # Bước 3: Kiểm tra nếu không tìm thấy số nào
    if not danh_sach_so:
        return 0, 0, 0 # Số lượng, Tổng, Trung bình đều là 0
    
    # Bước 4: Tính toán
    so_luong = len(danh_sach_so)
    tong = sum(danh_sach_so)
    trung_binh = tong / so_luong
    
    return so_luong, tong, trung_binh

# --- Chạy thử ---
doan_van = "Trong năm 2023, tôi đã đọc 12 cuốn sách và đi du lịch 5 lần."
sl, s, tb = xu_ly_so_trong_van_ban(doan_van)

print(f"Các số tìm được: {re.findall(r'\d+', doan_van)}")
print(f"Tổng các số: {s}")
print(f"Trung bình cộng: {tb:.2f}")
```

W7A27 
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ 
hay không 

```python
def kiem_tra_ngoac_hop_le(s):
    # Bước 1: Tạo "bản đồ" các cặp ngoặc
    # Khóa là dấu đóng, giá trị là dấu mở tương ứng
    mapping = {")": "(", "}": "{", "]": "["}
    
    # Bước 2: Chuẩn bị ngăn xếp (Stack)
    stack = []
    
    # Bước 3: Duyệt từng ký tự trong xâu
    for char in s:
        # Nếu là dấu đóng
        if char in mapping:
            # Lấy phần tử trên cùng của stack ra (nếu có)
            # Nếu stack rỗng, gán giá trị giả '#' để so sánh
            top_element = stack.pop() if stack else '#'
            
            # Kiểm tra xem dấu mở vừa lấy ra có khớp với dấu đóng hiện tại không
            if mapping[char] != top_element:
                return False
        else:
            # Nếu là dấu mở, đẩy vào stack
            stack.append(char)
            
    # Bước 4: Nếu stack rỗng thì là hợp lệ, ngược lại là dư thừa dấu mở
    return len(stack) == 0

# --- Chạy thử ---
test_cases = ["()[]{}", "([)]", "{[]}", "(("]

for test in test_cases:
    result = kiem_tra_ngoac_hop_le(test)
    print(f"Dãy '{test}': {'Hợp lệ' if result else 'Không hợp lệ'}")
        
    
```

Bà ̀ i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên 
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ 
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua) 

```python
def tim_day_hop_so_dai_nhat(n):
    if n < 4:
        return "Không có hợp số nào trong phạm vi này."

    # Bước 1: Sàng số nguyên tố để biết số nào là hợp số
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False
    for p in range(2, int(n**0.5) + 1):
        if is_prime[p]:
            for i in range(p * p, n + 1, p):
                is_prime[i] = False

    # Bước 2: Duyệt tìm dãy hợp số dài nhất
    max_len = 0
    start_index = -1
    
    current_start = -1
    current_len = 0
    
    for i in range(1, n + 1):
        # Điều kiện là hợp số: > 1 và không phải nguyên tố
        if i > 1 and not is_prime[i]:
            if current_start == -1:
                current_start = i
            current_len += 1
        else:
            # Gặp số nguyên tố hoặc số 1 -> Kết thúc một dãy hợp số
            if current_len > max_len:
                max_len = current_len
                start_index = current_start
            
            # Reset để tìm dãy tiếp theo
            current_start = -1
            current_len = 0
            
    # Kiểm tra lần cuối nếu dãy dài nhất nằm ở cuối mảng
    if current_len > max_len:
        max_len = current_len
        start_index = current_start

    # Trả về kết quả
    if start_index != -1:
        return list(range(start_index, start_index + max_len))
    return []

# --- Chạy thử với N = 20 ---
n_input = 20
ket_qua = tim_day_hop_so_dai_nhat(n_input)
print(f"Dãy hợp số dài nhất trong khoảng [1, {n_input}] là: {ket_qua}")
# Giải thích: 
# Primes: 2, 3, 5, 7, 11, 13, 17, 19
# Các dãy hợp số: [4], [6], [8,9,10], [12], [14,15,16], [18], [20]
# Dài nhất là [8,9,10] và [14,15,16]. Lấy dãy đầu tiên là [8,9,10].
```

```python
def kiemtra(n):
    def songuyento(s):
        for i in range(2,int(s**0.5)+1):
            if s % i == 0:
                return False
        return True
    lst = []
    lst1 = []
    for i in range(2, n + 1):
        if not songuyento(i):
            lst1.append(i)
        else:
            if len(lst1) > len(lst):
                lst = lst1
            lst1 = []
    if len(lst1) > len(lst):
        lst = lst1
    return lst

n = 20
print(f"Dãy hợp số dài nhất <= {n}: {kiemtra(n)}")


```

W7A30 
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím 

```python
def tim_lis_khong_lien_tuc(arr):
    if not arr:
        return []

    n = len(arr)
    # dp[i] lưu độ dài LIS kết thúc tại i
    dp = [1] * n
    # trace[i] lưu vị trí của phần tử đứng trước arr[i] trong dãy LIS
    trace = [-1] * n

    # Bước 2: Xây dựng bảng phương án (Quy hoạch động)
    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j]:
                if dp[i] < dp[j] + 1:
                    dp[i] = dp[j] + 1
                    trace[i] = j

    # Bước 3: Tìm vị trí có dp lớn nhất
    max_len = max(dp)
    last_index = dp.index(max_len)

    # Bước 4: Truy vết để lấy lại dãy số
    result = []
    while last_index != -1:
        result.append(arr[last_index])
        last_index = trace[last_index]
    
    # Vì truy vết từ cuối lên nên phải đảo ngược lại
    return result[::-1]

# Chạy thử
arr = [10,2,5,3]
ket_qua = tim_lis_khong_lien_tuc(arr)
print(f"Dãy con tăng dần không liên tục dài nhất là: {ket_qua}")
```
