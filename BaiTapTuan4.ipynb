Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)

```python
def songuyento(n):
    if n <2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
while True:
    try:
        line = input()
        if not line:
            break
        a, b = map(int, line.split())
        tong = 0
        for x in range(a, b + 1):
            if songuyento(x):
                tong += x
        print(tong)
        break
    except:
        print('nhập lại 2 số a, b')
        break
```

Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def dao_nguoc(n):
    s = str(n)
    s_dao = s[::-1]
    return int(s_dao)
def la_doi_xung(n):
    return n == dao_nguoc(n)
try:
    n = int(input())
    if n <= 0:
        print('Vui lòng nhập lại số n')
    else:
        buoc = 0
        while not la_doi_xung(n):
            n_dao = dao_nguoc(n)
            n = n + n_dao
            buoc += 1
        print(f'Số bước: {buoc}')
        print(f'Giá trị palindrome: {n}')
except ValueError:
    print('Lỗi: Bạn phải nhập một số nguyên')
```

Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
cách). Nếu không có, in ra (no number).

```python
def sochinhphuong(a):
    if a < 0:
        return False
    else:
        s = int(a**0.5)
        if s * s == a:
            return True
        else:
            return False
def chusodoimotkhacnhau(a):
    if a == 0:
        return True
    danhsach = [False] * 10
    temp  = a
    while temp > 0:
        b = temp % 10
        if danhsach[b] == True:
            return False
        danhsach[b] = True
        temp = temp // 10
    return True
dayso = []
n = int(input())
for a in range(1, n + 1):
    if sochinhphuong(a) and chusodoimotkhacnhau(a):
        dayso.append(a)
if len(dayso) == 0:
    print('No number')
else:
    for i in dayso:
        print(i, end=' ')
```

Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất." 

```python
def dodaicollatz(x):
    if x == 1:
        return 1
    tong = 1
    while x != 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = 3 * x + 1
        tong += 1
    return tong
try:
    n = int(input())
    if n < 1:
        print('Vui lòng nhâpj số nguyên dương')
    else:
        max_L = 0
        best_x = 0
        for i in range(1, n + 1):
            L = dodaicollatz(i)
            if L > max_L:
                max_L = L
                best_x = i
        print(f'Kết quả: {best_x} {max_L}')
except:
    print('Lỗi: Nhập lại đuê')
```

Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: 
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a, b = map(int, input('Nhập 2 số cần xét: ').split())
tonga = 0
tongb = 0
for i in range(1, a):
    if a % i == 0:
        tonga += i
for i in range(1, b):
    if b % i == 0:
        tongb += i
if tonga == b and tongb == a:
    print(f'Hai số {a}, {b} là cặp số thân thiết')
else:
    print(f'Hai số {a}, {b} không là cặp số thân thiết')
```

Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
tìm ước chung lớn nhất của hai số m và n.

```python
m, n = map(int, input('Nhập 2 só m và n cần xét vào đây: ').split())
danhsach = []
cacso = []
for i in range(1, m + 1 ):
    if m % i == 0:
        danhsach.append(i)
for i in danhsach:
    if n % i == 0:
        cacso.append(i)
if len(cacso) == 0:
    print(f'Hai số {m}, {n} không có ước chung')
else:
    L = 0
    for i in cacso:
        if i > L:
            L = i
    print(L)
```
