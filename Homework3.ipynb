
Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 
Ví dụ: Ta nhập h = 4: 

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    khoang_trang_dau = " " * (h - i)
    if i == 1:
        print(khoang_trang_dau + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        khoang_trang_giua = " " * (2 * i - 3)
        print(khoang_trang_dau + "*" + khoang_trang_giua + "*")
    
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 

```python
a = int(input())
if a < 0:
    print(f'{a} không là số chính phương')
else:
    can = int(a ** 0.5)
    if can * can == a:
        print(f'{a} là số chính phương')
    else:
        print(f'{a} không là số chính phương')
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
a = int(input())
tong = 0
for i in range(1, a):
    if a % i == 0:
        tong += i
if tong <= a:
    print(f'{a} là số hoàn hảo')
else:
    print(f'{a} không là số hoàn hảo')
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 


```python
n = int(input('Nhập 1 số bất kì để khởi động chương trình'))
if n == -1:
    print('(no number)')
else:
    max = n
    min = n
    while True:
        n = int(input('Bắt đầu nhập các giá trị vào đây'))
        if n == -1:
            break
        elif n > max:
            max = n
        elif n < min:
            min = n
print(f'Số lớn nhất trong dãy bạn vừa nhập vào là {max}')
print(f'Số bé nhất trong dãy bạn vừa nhập là {min}')
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều 
cách nhất có thể, ko dùng mảng.

```python
n = int(input())
tongcacchuso = 0
while True:
    s = n % 10
    if s != 0:
        tongcacchuso += 1
        n = n // 10
    else:
        break
print(tongcacchuso)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ 

```python
n = int(input())
tongchusochan = 0
tongchusole = 0
while True:
    s = n % 10
    if s == 0:
        break
    else:
        if s % 2 == 0:
            tongchusochan += 1
        else:
            tongchusole += 1
        n = n // 10
print(tongchusochan)
print(tongchusole)
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())
sotruoc = 1
sosau = 1
while sosau < A:
    sosau = sotruoc + sosau
    sotruoc = sosau - sotruoc
print(sotruoc)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
def kiemtrasonguyento(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int((n**0.5) + 1 ), 2):
        if n % i == 0:
            return False
    return True
n = int(input())
tong = 0
while n > 0:
    s = n % 10
    if kiemtrasonguyento(s):
        tong += 1
    n //= 10
print(tong)
```
