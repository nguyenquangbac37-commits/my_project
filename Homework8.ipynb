Xây dựng lớp Shape với phương thức area(). 
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area(). 
Lấy π = 3.14. 
 
Input 
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE) 
Dòng 2: tham số tương ứng. 
 
Output 
In ra diện tích với 2 chữ số thập phân.

```python
class Shape:
    def area(self):
        return 0
class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h 
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * (self.r ** 2)
loaihinh = input().strip().upper()
if loaihinh =='RECTANGLE':
    w, h = map(float, input().split())
    hinh = Rectangle(w, h)
elif loaihinh == 'CIRCLE':
    r = float(input())
    hinh = Circle(r)
print(f'hinh:.2f')
```

Input 
Một dòng chứa 3 số nguyên a b c. 
 
Output 
In ra chu vi tam giác.

```python
class Polygon():
    def perimeter(self):
        return 0
class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c
a, b, c = map(float, input().split())
print(Triangle(a,b,c).perimeter())
```

Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance(). 
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức 
add_interest() để cộng balance*rate vào số dư. 
 
Input 
balance rate 
 
Output 
Số dư sau khi cộng lãi, in 2 chữ số thập phân. 

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def get_balance(self):
        return self.balance 
class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate
    def add_interest(self):
        self.balance = self.balance + (self.balance * self.rate) 
myacc = SavingAccount(100, 0.05)
myacc.add_interest()
print(f"Số dư sau khi cộng lãi: {myacc.get_balance():.2f}")
```

Xây dựng lớp Printer với phương thức print_value(x). 
Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của 
số nguyên dương x. 
 
Input 
Một số nguyên dương n. 
 
Output 
In ra dạng nhị phân của n.

```python
class Printer:
    def print_value(self, x):
        return x
class BinaryPrinter(Printer):
    def print_value(self, x):
        danhsach = []
        while x >0:
            k = x % 2
            danhsach.append(str(k))
            x //= 2
        ketqua = ''.join(danhsach[::-1])
        return ketqua
n = int(input())
printer= BinaryPrinter()
print(printer.print_value(n))
        
```

Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận: 
[a b] 
[c d] 
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc. 
 
Input 
Một dòng chứa 4 số nguyên a b c d. 
 
Output 
In ra trace và determinant trên cùng một dòng. 

```python
class Matrix2x2:
    def __init__(self, a,b,c,d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        total = self.a + self.d
        return total
    def determinant(self):
        total1 = (self.a * self.d) - ( self.b * self.c)
        return total1
    def __str__(self):
        return f"[{self.a} {self.b}]\n[{self.c} {self.d}]"
a ,b, c, d = map(int, input().split())
ketqua = Matrix2x2(a, b, c, d)
print(ketqua.trace())
print(ketqua.determinant()) 
print(ketqua.__str__())
```

Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức 
final_price() để trả về giá sau giảm. 
Công thức: price * (100 - percent) / 100 
 
Input 
name price percent 
 
Output 
In ra giá cuối cùng với 2 chữ số thập phân.

```python
class Produce:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def final_price(self):
        return f'{self.price:.2f}'
class DiscountProduct(Produce):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent
    def final_price(self):
        return f'{self.price * ((100 - self.percent)/100):.2f}'
danhsach = list(map(str, input().split()))
ketqua = DiscountProduct(danhsach[0], int(danhsach[1], int(danhsach[2])))
print(ketqua.final_price())
```

Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), 
value(). 
Sau đó thực hiện dãy thao tác trên đối tượng. 
 
Input 
Dòng 1: n 
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách. 
 
Output 
Giá trị cuối cùng của count. 

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count
n = int(input().strip())
danhsachlenh = input().split()
maydem = Counter()
for lenh in danhsachlenh:
    lenhchuan = lenh.upper()
    if lenhchuan == 'INC':
        maydem.inc()
    elif lenhchuan =='DEC':
        maydem.dec()
print(maydem.value())
```

W8A30 Yêu cầu 
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), 
pop(), top(). 
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ. 
 
Input 
Dòng 1: n 
n dòng tiếp theo là một trong ba lệnh: 
PUSH x 
POP 
TOP 
 
Output 
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack:
    def __init__(self):
        self.items = []
    def push(self, x):
        self.items.append(x)
    def pop(self):
        if len(self.items) > 0:
            self.items.pop()
    def top(self):
        return self.items[-1]
n_str = input().strip()
if n_str:
    n = int(n_str)
    my_stack = Stack()
    for _ in range(n):
        command = input().split()
        type_cmd = command[0].upper()
        if type_cmd == 'PUSH':
            value = command[1]
            my_stack.push(value)
        elif type_cmd =='Top':
            my_stack.pop()
        elif type_cmd == 'TOP':
            print(my_stack.top())
        
```
