Kiểm tra xem một số có phải là lũy thừa của 2 hay không 
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.

```python
n = int(input())
if n > 0 and (n & (n -1)) ==0:
    print(f'{n} là lũy thừa của 2')
else:
    print(f'{n} không là lũy thừa của 2')
```

[Even and Odd]  
Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ. Nếu x là số chẵn, in ra "Even". Nếu x 
là số lẻ, in ra "Odd".  

```python
n = int(input('nhập số nguyên dường vào đây'))
if n >0 and n % 2 ==0:
    print('Even')
elif n > 0 and n % 2 != 0:
    print('Odd')
else:
    print(f'số {n} bạn nhập vào sai yêu cầu')

```

Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau: - 0-50 kWh: 1500đ/kWh - 51-100 kWh: 2000đ/kWh - 100 kWh: 3000đ/kWh 

```python
n = int(input('nhập số điện tiêu thụ'))
if 0 < n <= 50:
    tiendien = n * 1500
elif 51 <= n <= 100:
    tiendien = 50 * 1500 + (n - 50) * 2000
else:
    tiendien = 50 * 1500 + 50 * 2000 + (n - 100) * 3000
print(tiendien)
```

Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc: - >= 8.0: Giỏi - >= 6.5: Khá - >= 5.0: Trung bình - < 5.0: Yếu

```python
n = float(input())
if n >= 8:
    print('Giỏi')
elif 6.5 <= n < 8:
    print('Khá')
elif 5 <= n < 6.5:
    print('Trung bình')
else:
    print('Yếu')

```

Cấp số nhân 
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng 
thứ tự a, b, c, d hay không?

```python
a, b, c, d = map(float, input().split())
s = b / a
if c == b * s and d == c * s:
    print(f'các số {a}, {b}, {c}, {d} tạo thành cấp số nhân theo thứ tự')
else:
    print('Các số đã cho không tạo thành cấp số nhân')
```
