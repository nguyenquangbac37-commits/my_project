W6A17 
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
n = int(input())
matran = []
for i in range(n):
    hang = list(map(int, input().split()))
    matran.append(hang)
duongcheochinh = []
for i in range(n):
    duongcheochinh.append(matran[i][i])
duongcheophu = []
for i in range(n):
    duongcheophu.append(matran[i][n - 1 - i])
print(duongcheochinh)
print(duongcheophu)
```

W6A20 
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1 

```python
def kiemtra(list, k):
    tong = 0
    for i in range(len(list)):
        if list[i] == k:
            tong += 1
            if tong == 1:
                a = i + 1
                return a
    return -1
danhsach = list(map(int, input().split()))
k = int(input())
print(kiemtra(danhsach, k))
```

W6A21 
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
danh_sach = input("Nhập chuỗi: ").split()
temp_dict = {}
for index, word in enumerate(danh_sach):
    if word in temp_dict:
        temp_dict[word].append(index)
    else:
        temp_dict[word] = [index]
ket_qua = {}
for word, positions in temp_dict.items():
    ket_qua[word] = tuple(positions)
print(ket_qua)
```

W6A22 
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
trong list 

```python
def kiemtra(list):
    giatrimax = list[0]
    for i in list:
        if i > giatrimax:
            giatrimax = i
    cacvitri = []
    for i in range(len(list)):
        if list[i] == giatrimax:
            cacvitri.append(i + 1 )
    return giatrimax, cacvitri
danhsach = list(map(int, input().split()))
print(kiemtra(danhsach))
            
```

W6A28 
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
thực sự, hay vô danh (không tăng, không giảm) 

```python
danhsach = list(map(int, input().split()))
def daytang(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] > list[j]:
                list[i], list[j] = list[j], list[i]
    return list
def daygiam(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] < list[j]:
                list[i], list[j] = list[j], list[i]
    return list
def kiemtra(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] == list[j]:
                return True
    return False
if kiemtra(danhsach):
    print('Vô danh')
else:
    if danhsach == daytang(danhsach.copy()):
        print('Dãy tăng')
    elif danhsach == daygiam(danhsach.copy()):
        print('Dãy giảm')
    else:
        print('Vô danh')
```
