Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def amstronhg(n):
    tongchuso = 0
    while n % 10 != 0:
        tongchuso += 1
        n //= 10
    danhsach = []
    while n // 10 !=0:
        s = n % 10
        danhsach.append(s)
        n //= 10
    tong = 0
    for i in range(len(danhsach)):
        tong += i ** tongchuso
    if tong == n:
        return True
    return False
print(amstronhg(9474))
```

```python
def sokidieu(n):
    if n < 0:
        return False
    temp = n
    while temp > 10:
        tong = 0
        while temp % 10 != 0:
            tong += temp % 10
            temp //= 10
        temp = tong
    return temp == 1
print(sokidieu(18))
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 

```python
def detect_number_type(s):
    temp = s
    if s.startswith('+') or s.startswith('-'):
        temp = s[1:]
    if not temp:
        return 'neither'
    if temp.isdigit():
        return 'int'
    if '.' in temp:
        parts = temp.split('.')
        if len(parts) == 2:
            if parts[0].isdigit() and parts[1].isdigit():
                return 'float'
    return 'neither'
print(detect_number_type('1234'))
#### chữ số nhập vào sẽ ở dạng str nên đặt trong dấu ngoặc kép(đơn)
```

```python
def is_one_product_of_others(a, b, c):
    if a * b == c or a * c == b or b * c == a:
        return True
    return False
```

W5A25 Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1) 

```python
def point_in_quadrilateral(quad, p):
    xp, yp = p
    danhsach = []
    for i in range(4):
        p1 = quad[i]
        p2 = quad[(i+ 1) % 4]
        x1, y1 = p1
        x2, y2 = p2
        cp = (x2 - x1)*(yp - y1) - (y2 - y1)*(xp - x1)
        danhsach.append(cp)
    co_duong = any(i > 0 for i in danhsach)
    co_am = any(i <0 for i in danhsach)
    return not (co_duong and co_am)
quad = []
for i in range(4):
    diem = tuple(map(int, input().split()))
    quad.append(diem)
p = tuple(map(int, input().split()))
print(point_in_quadrilateral(quad, p))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 

```python
def tinhgoc(A, B, C):
    import math
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    #BA = x1 - x2, y1 - y2
    #BC = x3 - x2, y3 - y2
    s1 = ((x1-x2) * (x3-x2)) + ((y1 - y2)*(y3 - y2))
    s2 = (((x1-x2)**2 + (y1-y2)**2)**0.5)*(((x3-x2)**2 + (y3-y2)**2)**0.5)
    cosABC = s1 / s2
    cosABC = max(-1, min(1, cosABC))
    return math.degrees(math.acos(cosABC))
A = tuple(map(float, input().split()))
B = tuple(map(float, input().split()))
C = tuple(map(float, input().split()))
print(round(tinhgoc(A, B, C), 2))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k). 

```python
def nhithuc(n,k):
    giaithua1 = 1
    giaithua2 = 1
    giaithua3 = 1
    for i in range(1, n + 1):
        giaithua1 *= i
    for i in range(1, k+1):
        giaithua2 *= i
    for i in range(1, n-k+1):
        giaithua3 *= i
    return giaithua1 / (giaithua2 * giaithua3)
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, 
giá km, phí chờ, phụ phí ban đêm và giảm giá. 

```python
import math
def cuocphitaxi(car_type, start_hour, km, wait_minutes):
    #Bước 1 tính phí đi xe
    faremove = 0
    if car_type == 4:
        faremove = 11000
        if km > 0.8:
            if km < 20:
                faremove +=  (km -0.8) * 13000
            else:
                faremove += ( 20 - 0.8) * 13000 + (km -20) * 10500
    elif car_type == 7:
        faremove = 12000
        if km > 0.8:
            if km < 20:
                faremove += (km -0.8)* 14000
            else:
                faremove += (20 - 0.8) * 14000 + (km - 20) * 12000
    #tính phí chờ
    farewait = 0
    if wait_minutes > 5:
        blocks = math.ceil(wait_minutes / 5)
        farewait = (blocks - 1) * 2000
    #phụ phí ban đêm
    subtotal = faremove + farewait
    night_hours = [22, 23, 0, 1, 2, 3, 4, 5]
    is_night = start_hour in night_hours
    if is_night:
        subtotal *= 1.2
    if km > 30 and not is_night:
        subtotal *= 0.9
    total = int(subtotal)
    if total < 50000 and km > 5:
        total = 50000
    return total
car_type, start_hour, km, wait_minutes = map(int, input().split())
print(f'{cuocphitaxi(car_type, start_hour, km, wait_minutes)} VND')
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0.

```python
def khacnhau(n):
    danhsach = []
    while n % 10 != 0:
        s = n % 10
        danhsach.append(s)
        n //= 10
    for i in range(len(danhsach) - 1):
        for j in range(i + 1, len(danhsach)):
            if danhsach[i] == danhsach[j]:
                return False
    return True
def sum_1_to_n(n):
    if n <= 0:
        return 0
    else:
        tong = 0
        for i in range(1, n + 1):
            if khacnhau(i):
                tong += i
        return tong
print(sum_1_to_n(12))
```
