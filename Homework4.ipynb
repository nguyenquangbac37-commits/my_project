W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
temp = n
while n >= 2 and n % 2 == 0:
    n = n / 2
if n == 1:
    print(f'{temp} là một lũy thừa của 2')
else:
    print(f'{temp} không là một luỹ thừa của 2')
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input())
temp = n
tong = 0
while n // 10 != 0:
    i = n % 10
    tong += i
    n = n // 10
tong += n
print(f'Tổng các chữ số của {temp} là {tong}')
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
temp = n
tongchan = 0
tongle = 0
while n % 10 != 0:
    if (n % 10) % 2 == 0:
        tongchan += 1
    else:
        tongle += 1
    n = n // 10
print(f'{temp} có {tongchan} chữ số chẵn')
print(f'{temp} có {tongle} chữ số lẻ')
```

Cho S(k) = 1 + 2 + 3 + … + k

Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
danhsach = []
while True:
    n = int(input())
    if n == -1:
        break
    danhsach.append(n)
L = 0
for i in danhsach:
    if i > L:
        L = i
N = danhsach[0]
for i in danhsach:
    if i < N:
        N = i
print(f'Số lớn nhất là {L}')
print(f'Số bé nhất là {N}')
```

W4A26 Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A 

```python
n = int(input())
sotruoc = 1
sosau = 1
while sosau < n:
    sosau = sotruoc + sosau
    sotruoc = sosau - sotruoc
print(sotruoc)
```

W4A29 
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
thường và thường thành hoa. 

```python
n = input()
for i in n:
    a = ord(i)
    if 65 <= a <= 90 :
        i = chr( a + 32 )
    elif 97 <= a <= 122:
        i = chr( a - 32 )
    print(i, end='')
```
