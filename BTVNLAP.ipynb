### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
danhsachso = []
danhsach = []
while True:
    c = input('nhập số vào đây')
    if c == 'x' or c == "X":
        break
    conso = int(c)
    danhsachso.append(conso)
b = (len(danhsachso)-1)*(1/2)
for so in danhsachso:
    if so > b:
        danhsach.append(so)
print(len(danhsach))
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def chuyenkytu(x, n):
    if 'A' <= x <= 'Z':
        return chr((ord(x) - ord('A') + n) % 26 + ord('A'))
    elif 'a' <= x <= 'z':
        return chr((ord(x) - ord('a') + n) % 26 + ord('a'))
    else:
        return x
x = input('x là: ' )
n = int(input('bạn muốn tiến sau bao nhiêu kí tự: '))
ketqua = chuyenkytu(x, n)
print(ketqua)
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
n = int(input())
for i in range(1, n +1):
    print('*' * i)
```
